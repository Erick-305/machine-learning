### Normalizacion

```python
# EJERCICIO 1: NORMALIZACIÓN (MIN-MAX)

# Datos de ejemplo (por ejemplo, edades)
datos = [10, 20, 30, 40, 50]

# Paso 1: encontrar el mínimo y máximo
min_val = min(datos)
max_val = max(datos)

# Paso 2: aplicar la fórmula de normalización
# Fórmula: (x - min) / (max - min)

datos_normalizados = []

for x in datos:
    valor_normalizado = (x - min_val) / (max_val - min_val)
    datos_normalizados.append(valor_normalizado)

# Mostrar resultados
print("Datos originales:", datos)
print("Datos normalizados:", datos_normalizados)
```

### Estandarizacion

```python
# EJERCICIO 2: ESTANDARIZACIÓN (Z-SCORE)

import numpy as np

# Datos de ejemplo (por ejemplo, notas)
datos = [60, 70, 80, 90, 100]

# Paso 1: calcular la media y desviación estándar
media = np.mean(datos)
desviacion = np.std(datos)

# Paso 2: aplicar la fórmula de estandarización
# Fórmula: (x - media) / desviación estándar

datos_estandarizados = []

for x in datos:
    valor_estandarizado = (x - media) / desviacion
    datos_estandarizados.append(valor_estandarizado)

# Mostrar resultados
print("Datos originales:", datos)
print("Media:", media)
print("Desviación estándar:", desviacion)
print("Datos estandarizados:", datos_estandarizados)
```
